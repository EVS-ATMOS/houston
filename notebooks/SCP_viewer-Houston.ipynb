## Statistical Coverage Product viewer

This notebook makes timeseries plots of Peter May's Statistical Coverage Product as generated by derive_scp.py

```python
import xarray
from matplotlib import pyplot as plt
from datetime import datetime
from matplotlib import dates
import numpy as np
%matplotlib inline
```

Input data path

```python
scp_netcdf_file_path = '/home/rjackson/blues_home/houston/code/SCP_Houston201505.cdf'
```

Load netCDF variables

```python
Dataset = xarray.open_mfdataset(scp_netcdf_file_path)
Dataset = Dataset.sortby('time', ascending=True)
time = Dataset.time
SCP = Dataset.SCP
z = Dataset.z
print(len(z.values))
time = time.sortby('time')
SCP0 = SCP.isel(reflectivity=0)
SCP10 = SCP.isel(reflectivity=2)
SCP20 = SCP.isel(reflectivity=4)
SCP30 = SCP.isel(reflectivity=7)
```

Display statistical coverage product

```python
date_to_plot = '2015-5-20' 
plt.figure(figsize=(8,8))
xarray.plot.pcolormesh(SCP0.sel(time=date_to_plot), x='time', y='z',
                       cmap='terrain', vmin=0, vmax=100, size=8)
plt.title(('% Z > 0 dBZ ' + date_to_plot))
```

```python
date_to_plot = '2015-5-20'
plt.figure(figsize=(8,8))
xarray.plot.pcolormesh(SCP10.sel(time=date_to_plot), x='time', y='z',
                       cmap='terrain', vmin=0, vmax=100, size=8)
plt.title(('% Z > 10 dBZ ' + date_to_plot))
```

```python
date_to_plot = '2015-5-20'
plt.figure(figsize=(8,8))
xarray.plot.pcolormesh(SCP30.sel(time=date_to_plot), x='time', y='z',
                       cmap='terrain', vmin=0, vmax=5, size=8)
plt.title(('% Z > 35 dBZ ' + date_to_plot))
```
